```python
import os
import shutil
import random

def select_random_files(src_folder, num_files_per_class, dest_folder):
    # Get all folders in the source directory
    class_folders = [f for f in os.listdir(src_folder) if os.path.isdir(os.path.join(src_folder, f))]
    
    # Create destination folders
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    for class_folder in class_folders:
        class_path = os.path.join(src_folder, class_folder)
        files = [os.path.join(class_path, f) for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]

        # Check if there are enough files in the class folder
        if len(files) < num_files_per_class:
            raise ValueError(f"Not enough files in {class_folder} to select {num_files_per_class} files.")

        # Randomly shuffle and select files
        random.shuffle(files)
        selected_files = files[:num_files_per_class]

        # Create class folders in the destination directory
        dest_class_folder = os.path.join(dest_folder, class_folder)
        if not os.path.exists(dest_class_folder):
            os.makedirs(dest_class_folder)

        # Move selected files to the new folders
        for file_path in selected_files:
            shutil.copy(file_path, dest_class_folder)

def main():
    src_folder = 'tomato/New Plant Diseases Dataset(Augmented)/train'  # Replace with your source folder path
    dest_folder = 'mod_tomato_dataset_1'  # Replace with your destination folder path
    num_files_per_class = 150  # Number of files to select per class

    select_random_files(src_folder, num_files_per_class, dest_folder)
    print(f"Selected {num_files_per_class} files per class and saved them to {dest_folder}.")

if __name__ == '__main__':
    main()
```

```python

import os
import shutil
import random
import pandas as pd
import json

# Define paths
csv_file = 'Data/train.csv'  # CSV file containing image IDs and labels
json_file = 'Data/label_num_to_disease_map.json'  # JSON file mapping labels to disease names
dataset_path = 'train_images'  # Directory containing all the images
output_path = 'new_mod_cassava_dataset_1_modified'  # Directory where class folders will be created

# Create the output directory if it doesn't exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Read the CSV file
df = pd.read_csv(csv_file)

# Read the JSON file containing label-to-disease mapping
with open(json_file, 'r') as f:
    label_to_disease = json.load(f)

# Create folders for each class/disease
for label, disease in label_to_disease.items():
    class_folder = os.path.join(output_path, disease)
    if not os.path.exists(class_folder):
        os.makedirs(class_folder)

# Group the images by label
grouped = df.groupby('label')

# Process each class
for label, group in grouped:
    # Get the disease name from the label
    disease = label_to_disease[str(label)]
    
    # Sample 300 images from this class
    sampled_images = group.sample(300)['image_id'].tolist()
    
    # Define the path to the class folder
    class_folder = os.path.join(output_path, disease)
    
    # Copy the sampled images to the class folder
    for image_id in sampled_images:
        src = os.path.join(dataset_path, image_id)
        
        # Check if the image file exists before copying
        if os.path.exists(src):
            dst = os.path.join(class_folder, image_id)
            shutil.copy(src, dst)
        else:
            print(f"Image {image_id} not found, skipping...")

print("Images have been successfully extracted and sorted into class folders, skipping missing images.")
```

```python
import os
import shutil

# Define the paths
tomato_folder = 'mod_tomato_dataset_1'
cotton_folder = 'cotton_dataset_1_modified'
cassava_folder = 'new_mod_cassava_dataset_1_modified'
destination_folder = '(3400)renew_Modified_combined_dataset_cassava_tomato_cotton(3400)'

# List of source folders
source_folders = [tomato_folder, cotton_folder, cassava_folder]

# Create the destination folder if it does not exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Function to copy subdirectories from a source folder to the destination folder
def copy_subfolders(src_folder, dest_folder):
    for item in os.listdir(src_folder):
        item_path = os.path.join(src_folder, item)
        if os.path.isdir(item_path):
            # Skip .ipynb_checkpoints directories
            if item == '.ipynb_checkpoints':
                print(f"Skipping .ipynb_checkpoints in {src_folder}")
                continue
            
            dest_path = os.path.join(dest_folder, item)
            # Check if the destination path already exists
            if not os.path.exists(dest_path):
                # Copy the directory to the destination folder
                shutil.copytree(item_path, dest_path)
            else:
                print(f"Folder '{item}' already exists in '{dest_folder}' and will not be copied.")

# Iterate over each source folder and copy its subdirectories
for folder in source_folders:
    copy_subfolders(folder, destination_folder)

print("All subfolders have been combined successfully.")
```

```python
#Import the required modules
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.applications.vgg19 import VGG19


# Create the base model from the pre-trained model VGG19
base_model = VGG19(weights='imagenet', include_top=False)

# Add a global spatial average pooling layer
x = base_model.output
x = GlobalAveragePooling2D()(x)

# Add a fully-connected layer
x = Dense(1024, activation='relu')(x)

# Add a logistic layer with the number of classes in your dataset
predictions = Dense(20, activation='softmax')(x) # 20 classes

# this is the model we will train
vgg19_model = Model(inputs=base_model.input, outputs=predictions)

# first: train only the top layers (which were randomly initialized)
# i.e. freeze all convolutional VGG19 layers
for layer in base_model.layers:
    layer.trainable = False

# compile the model (should be done *after* setting layers to non-trainable)
vgg19_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image size and batch size
image_size = (128, 128)   # VGG19 requires 128*128 input
batch_size = 32           # Number of images to process in a batch

# Define the train data generator with augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,           # Rescale pixel values to [0, 1]
    shear_range=0.2,          # Apply random shear transformations
    zoom_range=0.2,           # Apply random zoom
    horizontal_flip=True,      # Randomly flip images horizontally
    validation_split=0.2      # Split 20% of the data for validation
)

# Define the test/validation data generator (only rescale)
test_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Set dataset directory path
dataset_path = '(340)renew_Modified_combined_dataset_cassava_tomato_cotton(340)'

# Create train data generator
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,      # Resize all images to 128*128
    batch_size=batch_size,
    class_mode='categorical',    # For multi-class classification
    subset='training',           # Use this for the training data
    shuffle=True                 # Shuffle the data for better training
)

# Create validation data generator
validation_generator = test_datagen.flow_from_directory(
    dataset_path,
    target_size=image_size,      # Resize all images to 128*128
    batch_size=batch_size,
    class_mode='categorical',    # For multi-class classification
    subset='validation',         # Use this for the validation data
    shuffle=False                # No need to shuffle the validation data
)

# Check the generators
print(f"Train Generator: {len(train_generator)} batches")
print(f"Validation Generator: {len(validation_generator)} batches")
```
